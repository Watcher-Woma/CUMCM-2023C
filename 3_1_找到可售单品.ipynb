```python
import os
import pandas as pd


# 指定文件夹路径
folder_path = "中间数据集\\多表_按单品"

# 初始化一个空列表来存储文件名
file_list = []

# 遍历文件夹中的所有文件
for root, dirs, files in os.walk(folder_path):
    for file in files:
        # 将文件名添加到列表中
        file_list.append(os.path.join(root, file))
        
file_list
```
