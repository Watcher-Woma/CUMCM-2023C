````python
import pandas as pd
import matplotlib.pyplot as plt # 基础绘图库（什么都能画，但可能慢、复杂）
import seaborn as sns # 数据可视化专用库（专门用来快速画数据图）


# 中文字体宋体，英文字体新罗马
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['font.serif'] = ['Times New Roman']
# 字体大小
plt.rcParams['font.size'] = 12  
# 正常显示负号
plt.rcParams['axes.unicode_minus'] = False
````

````python
def time_deal(df):

    # 转换时间列编码
    df["日期"] = pd.to_datetime(df["日期"])

    # 将时间列设为index
    df.set_index("日期", inplace=True)

    # 按月聚合
    df_finished = df.resample("M").sum()
    
    return df_finished
````

[['日期', '销量(千克)']]这部分代码表示在读取CSV文件后，只选择DataFrame中的'日期'和'销量(千克)'这两列数据。这通过选择DataFrame的特定列来创建一个新的DataFramedf1，只包含所需的日期和销量信息。

````python
df1 = pd.read_csv('中间数据集\\多表_按品类\\花菜类.csv')[["日期", "销量(千克)"]]
df1 = time_deal(df1)
df1.head()
````

# 描述型统计分析

```
这个函数首先从DataFrame（df）中排除对象类型（通常是分类数据或字符串）的列，然后对剩余的数值列计算一系列描述性统计量。这些统计量包括均值、四分位数（25%、50%和75%）、四分位距（IQR）和众数。结果被组织成一个新的DataFrame返回，以便用户可以方便地查看和分析这些统计信息。

使用这个函数的好处是它提供了一个简洁的方法来计算自定义的一组统计量，而不仅仅是依赖Pandas的describe()方法提供的默认统计信息。这对于需要特定统计分析的数据科学项目或研究非常有用。
```

````python
def describe(df):
    # 删除分类特征列
    df = df.select_dtypes(exclude=['object']) #  选择数据框中所有非object类型的列，即只保留数值型列

    # 计算统计量
    stats_df = { #  创建一个字典，用于存储DataFrame的各种统计量
        '均值': df.mean(), #  计算每列的平均值
        '四分位数1': df.quantile(0.25), #  计算每列的第一四分位数(25%分位数)
        '中位数': df.median(), #  计算每列的中位数(50%分位数)
        '四分位数3': df.quantile(0.75), #  计算每列的第三四分位数(75%分位数)
        '四分位距': df.quantile(0.75) - df.quantile(0.25), #  计算四分位距(第三四分位数与第一四分位数的差)
        '众数': df.mode().iloc[0],   #  计算每列的众数(取第一个众数值，因为可能有多个众数)
        '标准差': df.std(), #  计算每列的标准差
        '方差': df.var(), #  计算每列的方差
        '偏度': df.skew(), #  计算每列的偏度(衡量数据分布的不对称性)
        '峰度': df.kurtosis(), #  计算每列的峰度(衡量数据分布的尾部厚度)
    }

    # 创建一个新的DataFrame用于存储统计量
    stats_df = pd.DataFrame(stats_df) #  将传入的stats_df转换为pandas DataFrame格式

    # 转置以便每个统计量一行
    stats_df = stats_df.T #  对统计数据进行转置操作，将行变为列，列变为行

    # 把stats_df粘贴都剪切板
    stats_df.to_clipboard(excel=True) #  将统计结果复制到剪贴板，excel=True表示以Excel兼容的格式复制

    return stats_df #  返回统计结果数据框

# 描述型统计分析：计算统计量
res_1 = describe(df1) #  对数据框df1进行描述性统计分析
res_2 = describe(df2) #  对数据框df2进行描述性统计分析
res_3 = describe(df3) #  对数据框df3进行描述性统计分析
res_4 = describe(df4) #  对数据框df4进行描述性统计分析
res_5 = describe(df5) #  对数据框df5进行描述性统计分析
res_6 = describe(df6) #  对数据框df6进行描述性统计分析
````

````python
df_ = pd.DataFrame({
    "花菜类": df1["销量(千克)"],
    "花叶类": df2["销量(千克)"],
    "茄类": df3["销量(千克)"],
    "辣椒类": df4["销量(千克)"],
    "食用菌": df5["销量(千克)"],
    "水生根茎类": df6["销量(千克)"]
    })
df_.head()
````

````python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 6)) #  导入必要的库并创建图形和坐标轴

#  使用热力图展示数据集的Spearman相关系数矩阵 
# square=True: 使热力图的每个单元格为正方形 
# cmap="OrRd": 使用橙红色调的颜色映射 
# annot=True: 在单元格中显示相关系数值 
# xticklabels和yticklabels: 设置x轴和y轴的标签为数据框的列名
# .corr()：这是DataFrame对象的一个方法，用于计算列之间的相关性系数矩阵
# df.columns是Pandas中DataFrame对象的一个属性，用于获取或设置DataFrame中的列名
sns.heatmap(data=(df_.corr(method="spearman")),  
            square=True, cmap="OrRd", annot=True, ax=ax,
            xticklabels=df_.columns, yticklabels=df_.columns)

ax.set_title('蔬菜品类销量相关系数矩阵', fontsize=14) #  设置图表标题和字体大小
plt.yticks(rotation=0) #  设置y轴刻度标签不旋转（保持水平）

fig.savefig('图片\\1_1_蔬菜品类销量相关系数矩阵.eps') #  将图形保存为EPS格式文件，保存在"图片"文件夹下
````
